# The Perils of Data Journalism: Amateurism, Ambiguity, and the Pursuit of Contrarianism

#### [Brian Keegan, Ph.D.](www.brianckeegan.com) \([@bkeegan](https://twitter.com/bkeegan)\) College of Humanities and Social Sciences, Northeastern University 

A spectre haunting is haunting the web --- the spectre of "data journalism." All of the powers of old Journalism have entered into holy evangelism to praise this paradigm: *Post* and *Times*, Silver and Klein, New York designers and Bay Area developers.

Where I depart in this "manifesto" is to echo others' observations that "data journalism", having faced the world and published their analyses, has been found wanting. I have found three specific critiques, made in part by Krugman \[[1](http://krugman.blogs.nytimes.com/2014/03/18/sergeant-friday-was-not-a-fox/),[2](http://krugman.blogs.nytimes.com/2014/03/18/further-thoughts-on-hedgehogs-and-foxes/),[3](http://krugman.blogs.nytimes.com/2014/03/23/tarnished-silver/),[4](http://krugman.blogs.nytimes.com/2014/03/26/data-as-slogan-data-as-substance/)\] and Noah Smith \[[5](http://noahpinionblog.blogspot.com/2014/03/data-buzzword-vs-data-actual-thing.html)\] (and abetted by Silver's [petulant response to Krugman](http://fivethirtyeight.com/datalab/for-columnist-a-change-of-tone/) and Kieran Healy's [excellent analysis](http://kieranhealy.org/blog/archives/2014/03/26/silver-vs-krugman/) of the same), the most persuasive:

1. **Amateurism.** Data journalists' exploratory analyses often overlook scientists', analysts', and other experts' well-established methods and hard-won expertise. That the data and tools for analyzing them are increasingly available is stupendous gain for science and society --- but with great power comes great responsibility. Just because the conclusions of "experts" in political media are terrifyingly unreliable does not mean the conclusions of "experts" in economics, climate science, and other scientific fields are similarly disruptable by enterprising data munging. Rather than attempting to become data analysts themselves, data journalists should perform preliminary analyses and seek the advice of or collaborate directly with well-credentialed scientists and analysts to corroborate them.

2. **Ambiguity.** These analyses appear with the veneer of empirical rigor, clad in infographics crying out for Edward Tufte's approval and containing intimidating words like "Bayesian" or "regression". But they offer little substantive detail in terms of the sources of these data and even fewer specifics about the types of analyses used. In scientific research, these methodological details---while inevitably dense and boring---are the instruction manual for interpreting the validity the analysis and its findings; you can do far [more damage with a bad model](http://xkcd.com/605/) than with an improperly-assembled [IKEA Björnholmen storage unit](http://www.buzzfeed.com/saraboboltz/some-people-arent-very-good-at-ikea-af7c). It is essential that journalists make the data and models they use open for inspection. This not only aids other researchers who want to replicate or extend the finding, it also compels the author to revise their information (how Bayesian!) and learn about new approaches.

3. **Pursuit of contrarianism.** I am very much sympathetic to journalists whose professional fortunes are increasingly yoked to the capricious tyrants of pageviews, optimized headlines, and other behavioral nebulae. However, the thirst for clicks has corrupted journalists' skepticism and contributed to an increasingly toxic culture of contrarianism. Several FiveThirtyEight articles have bear-hugged the framing of "everything you know about X is wrong." If everything that is known also true? [No](http://i.imgur.com/j74SykU.gif). Should we use data analysis to test claims? [Yes](http://xkcd.com/552/). Can data be tortured into making false confessions? [Easily](https://archive.org/details/HowToLieWithStatistics). But journalists, with correlations in hand from the most easily-scraped database, are not heroes for standing athwart scientific consensus, yelling stop.

### The Bechdel Test article in FiveThirtyEight

Walk Hickey published an article on April 1 on FiveThirtyEight, titled [The Dollar-And-Cents Case Against Hollywood’s Exclusion of Women](http://fivethirtyeight.com/features/the-dollar-and-cents-case-against-hollywoods-exclusion-of-women/). The article makes two central claims:

> 1. We found that the median budget of movies that passed the test...was substantially lower than the median budget of all films in the sample.

> 2. ...we found evidence that films that feature meaningful interactions between women may in fact have a better return on investment, overall, than films that don’t.

I call Claim 1 the "Budgets Differ" finding and Claim 2 the "Results Differ" finding.

The claims about the return on investment are particularly problematic both from a perspective of the claims made from the models as well as the how the statistics were estimated (emphasis added).
> We did a statistical analysis of films to test two claims: first, that films that pass the Bechdel test — featuring women in stronger roles — see a lower return on investment, and second, that they see lower gross profits. We found no evidence to support either claim.

> On the first test, we ran a regression to find out if passing the Bechdel test corresponded to lower return on investment. Controlling for the movie’s budget, which has a negative and significant relationship to a film’s return on investment, **passing the Bechdel test had no effect on the film’s return on investment.** In other words, adding women to a film’s cast didn’t hurt its investors’ returns, contrary to what Hollywood investors seem to believe.

> The total median gross return on investment for a film that passed the Bechdel test was \$2.68 for each dollar spent. The total median gross return on investment for films that failed was only \$2.45 for each dollar spent.

> On the second test, we ran a regression to find out if passing the Bechdel test corresponded to having lower gross profits — domestic and international. Also controlling for the movie’s budget, which has a positive and significant relationship to a film’s gross profits, once again **passing the Bechdel test did not have any effect on a film’s gross profits.**

Hickey concludes:

> ...our data demonstrates that films containing meaningful interactions between women **do better at the box office than movies that don’t**...

### Simple linear extrapolation

We'll begin with a simple linear model of the change in the average Bechdel scores over time.

```python
avg_bechdel = df.set_index('Released_x').groupby(lambda x:x.year)['new_rating'].agg(np.mean).ix[1912:]
avg_bechdel.index.name = 'date'
avg_bechdel.name = 'bechdel'
avg_bechdel = avg_bechdel.reset_index()
```

```python
l_model = smf.ols(formula='bechdel ~ date', data=avg_bechdel).fit()
print l_model.summary()
```

```python
# Create a DataFrame "X" that we can extrapolate into
X = pd.DataFrame({"date": np.linspace(start=1912.,stop=2112.,num=201)})

# Plot the observed data
plt.plot(avg_bechdel['date'],avg_bechdel['bechdel'],c='g',label='Observed')

# Plot the predictions from the model
plt.plot(X,l_model.predict(X),c='r',label='Linear model',lw=2)
plt.legend(loc='lower right')
plt.ylim((0,3))
plt.xlim((1912,2112))
plt.xlabel('Year')
plt.ylabel('Avg. Bechdel Test')

plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

Re-arrange the `y=Bx+I` equation to solve for `y=30`: `x=(y-I)/B`

Then convert the float into a proper date and return the day of the week as well (Mondays are 0).

```python
year_est = (3 - l_model.params['Intercept'])/l_model.params['date']
year_int = int(year_est)
d = datetime.timedelta(days=(year_est - year_int)*365.25)
day_one = datetime.datetime(year_int,1,1)
date = d + day_one
print date, date.weekday()
```

## Relationship between Bechdel score and revenue

This is the "raw" distribution of the relationship between Bechdel scores (x-axis) and annual revenue (y-axis). We want to model this relationship, but first we want to explain away other potentially confounding sources of variation in the relationship coming from differences in genre, critics' scores, fans' scores, production budgets, year of release, month of release, length of the movie, and opening theaters, among others. 

You'll notice each cloud of points around 0, 0.5, 1, 1.5, 2, 2.5, and 3. These are simply "jittered" by adding a bit of normally-distributed errors to the plots (but not the underlying data we're estimating) to show the frequency of datapoints without them all sitting on top of each other.

```python
# Add a jitter so the points aren't overlapping
x = df['new_rating'].apply(lambda x:float(x)+np.random.normal(0, 0.05))
y = df['Adj_Revenue']+1

# Plot with an alpha so the overlaps reveal something about relative density
plt.scatter(x,y,alpha=.2,label='Data',c='g')
plt.xticks(arange(0,4))
plt.yscale('log')
plt.autoscale()
plt.xlabel('Bechdel score',fontsize=18)
plt.ylabel('Adj. Revenue (2014 $)',fontsize=18)
plt.ylim((2e0,1e10))
plt.grid(False,which='minor')
#plt.legend(loc='right',bbox_to_anchor=(1.275,.5))
```

There's no clear trend here, lots of clouds of points with their bunches pretty much all at the same level, just below $10^8$. Let's run the numbers to see if there's actually a trend.

The first model we estimate is replicating the 538 analysis whether there's a relationship between revenue and Bechdel score, controlling for budget. We fit this model properly using `rating` as a categorical variable and logging the highly-skewed `Revenue` and `Budget` variables.

```python
m1 = smf.ols(formula='log(Adj_Revenue+1) ~ C(new_rating) + log(Adj_Budget+1)', data=df[df['Adj_Revenue']>0]).fit()
print m1.summary()
```

```python
m1_b0 = np.exp(m1.params['Intercept'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b1 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.1.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b2 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.2.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b3 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.3.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))

print "Movies with an average budget and passing no element of the Bechdel test will make ${:,} on average".format(round(m1_b0,-3))
print "Movies with an average budget and passing the first element of the Bechdel test will make ${:,} on average".format(round(m1_b1,-3))
print "Movies with an average budget and passing the second element of the Bechdel test will make ${:,} on average".format(round(m1_b2,-3))
print "Movies with an average budget but passing all elements of the Bechdel test will make ${:,} on average".format(round(m1_b3,-3))
```

Next we estimate a "bad" model using `rating` as a continuous variable so that we can visualize the model on top of the data itself.

```python
m1_bad_res = smf.ols(formula='log(Adj_Revenue+1) ~ new_rating + log(Adj_Budget+1)', data=df[df['Adj_Revenue']>0]).fit()
print m1_bad_res.summary()
```

```python
# Estimate model without the Bechdel test IV and containing only the IVs to get residulals
m1_bad_res = smf.ols(formula='log(Adj_Revenue+1) ~ log(Adj_Budget+1)', data=df[df['Adj_Revenue']>0]).fit()

# Put these residual estimates in a temporary DataFrame
df_temp = df
df_temp['residuals'] = m1_bad_res.resid

# Estimate a bad model with a continuous Bechdel rating against the residuals
m1_bad = smf.ols(formula='residuals ~ new_rating', data=df_temp).fit()
m1_bad_x = pd.DataFrame({'new_rating':np.linspace(min(df['new_rating']),max(df['new_rating']),100)})

# Add a jitter so the points aren't overlapping
x = df['new_rating'].apply(lambda x:float(x)+np.random.normal(0, 0.05))
y = df_temp['residuals']

plt.plot(m1_bad_x['new_rating'],m1_bad.predict(m1_bad_x),c='r',lw=2,label='Model')

# Plot with an alpha so the overlaps reveal something about relative density
plt.scatter(x,y,alpha=.2,label='Data',c='g')
plt.xticks(arange(0,4))
#plt.yscale('log')
plt.autoscale()
plt.xlabel('Bechdel score',fontsize=18)
plt.ylabel('Residual Adj. Revenue',fontsize=18)
plt.ylim((-10,10))
plt.legend(loc='lower left')
print "Every additional Bechdel 'point' adds {0} to the revenue and {1}".format(round(m1_bad.params['new_rating'],2),sig_tester(m1_bad.pvalues['new_rating']))
```
