```python
def sig_tester(num):
    if num<0.05:
        return 'is significant'
    else:
        return 'is NOT significant'
```

# Write scraper for Bechdel Test API

Documentation here: http://bechdeltest.com/api/v1/doc#getMovieByImdbId

First, download a list of all movie IDs using hte `getAllMovieIds` method. Inspect the first 5. Create a list of IMDB ids out of this list of dictionaries to pass to the API.

```python
movie_ids = json.loads(urllib2.urlopen('http://bechdeltest.com/api/v1/getAllMovieIds').read())
imdb_ids = [movie[u'imdbid'] for movie in movie_ids]
print u"There are {0} movies in the Bechdel test corpus".format(len(imdb_ids))
```

Pull an example `getMovieByImdbId` from the API.

```python
json.loads(urllib2.urlopen('http://bechdeltest.com/api/v1/getMovieByImdbId?imdbid=0000091').read())
```

Pull all of the data we can down. This may take a while. We don't want to have to repeat that crawl again, so let's write the data to disk as `bechdel.json`

Read the data back from disk and convert to `ratings_df`.

```python
ratings_df = pd.read_json('bechdel.json')
ratings_df['imdbid'] = ratings_df['imdbid'].dropna().apply(int)
ratings_df = ratings_df.set_index('imdbid')
ratings_df.tail()
```

```python
ratings_df['new_rating'] = ratings_df['rating']
ratings_df.loc[((ratings_df['dubious'] == u'0') & (ratings_df['rating'] == 0)),'new_rating'] = 0.0
ratings_df.loc[((ratings_df['dubious'] == u'1') & (ratings_df['rating'] == 1)),'new_rating'] = 0.5
ratings_df.loc[((ratings_df['dubious'] == u'0') & (ratings_df['rating'] == 1)),'new_rating'] = 1.0
ratings_df.loc[((ratings_df['dubious'] == u'1') & (ratings_df['rating'] == 2)),'new_rating'] = 1.5
ratings_df.loc[((ratings_df['dubious'] == u'0') & (ratings_df['rating'] == 2)),'new_rating'] = 2.0
ratings_df.loc[((ratings_df['dubious'] == u'1') & (ratings_df['rating'] == 3)),'new_rating'] = 2.5
ratings_df.loc[((ratings_df['dubious'] == u'0') & (ratings_df['rating'] == 3)),'new_rating'] = 3.0
```

They field we want is `rating` which is defined as: 

    "The actual score. Number from 0 to 3.
    0.0 means no two women,
    0.5 means two women (dubious),
    1.0 means no talking, 
    1.5 means two women talking (dubious), 
    2.0 means talking about a man, 
    2.5 means talking about something other than a man (dubious),
    3.0 means it passes the test."

# Pull down IMDB data

Pull an example from http://www.omdbapi.com/.

```python
json.loads(urllib2.urlopen('http://www.omdbapi.com/?i=tt0000091').read())
```

Crawl the data. This will take a while. Write the data to disk when you're done as `imdb_data.json`.

Read the data back from disk and convert to `imdb_df`. Print out an example row, but transpose it because there are a lot of columns.

```python
imdb_df = pd.read_json('imdb_data.json')
imdb_df['Released'] = pd.to_datetime(imdb_df['Released'], format="%d %b %Y", coerce=True)
imdb_df['imdbID'] = imdb_df['imdbID'].str.slice(start=2)
```

More cleanup happens below.

```python
# Remove the " min" at the end of Runtime entries so we can convert to ints
imdb_df['Runtime'] = imdb_df['Runtime'].str.slice(stop=-4).replace('',np.nan)

# Some errant runtimes have "h" in them. Commented-out code below identifies them.
#s = imdb_df['Runtime'].dropna()
#s[s.str.contains('h')]

# Manually recode these h-containing Runtimes to minutes
imdb_df.ix[946,'Runtime'] = '169'
imdb_df.ix[1192,'Runtime'] = '96'
imdb_df.ix[1652,'Runtime'] = '80'
imdb_df.ix[2337,'Runtime'] = '87'
imdb_df.ix[3335,'Runtime'] = '62'

# Convert Release into years
imdb_df['Year'] = imdb_df['Released'].apply(lambda x:x.year)

# Convert the series to float
imdb_df['Runtime'] = imdb_df['Runtime'].apply(float)

# Take the imdbVotes formatted as string containing "N/A" and comma-delimited thousands, convert to float
imdb_df['imdbVotes'] = imdb_df['imdbVotes'].dropna().replace('N/A',np.nan).dropna().apply(lambda x:float(x.replace(',','')))

# Take the Metascore formatted as string containing "N/A", convert to float
imdb_df['Metascore'] = imdb_df['Metascore'].dropna().replace('N/A',np.nan).dropna().apply(float)

# Take the imdbRating formatted as string containing "N/A", convert to float
imdb_df['imdbRating'] = imdb_df['imdbRating'].dropna().replace('N/A',np.nan).dropna().apply(float)

# Convert imdb_ID to int, set it as the index
imdb_df['imdbID'] = imdb_df['imdbID'].dropna().apply(int)
imdb_df = imdb_df.set_index('imdbID')
```

# Join data together

```python
df = imdb_df.join(ratings_df[['new_rating','title']].dropna(),how='inner').reset_index()
df = pd.merge(df,numbers_df,left_on='Title',right_on='Movie',how='inner')

adjusted_revenue = {index:revenue * inflator[year] for index,year,revenue in zip(list(df.index),list(df['Year_x']),list(df['Revenue'])) if year in range(1913,2015)}
adjusted_budget = {index:budget * inflator[year] for index,year,budget in zip(list(df.index),list(df['Year_x']),list(df['Budget'])) if year in range(1913,2015)}

df['Adj_Revenue'] = pd.Series(adjusted_revenue.values(),index=adjusted_revenue.keys())
df['Adj_Budget'] = pd.Series(adjusted_budget.values(),index=adjusted_budget.keys())

del df['Released_y'],df['Year_y']

df.rename(columns={'Year_x':'Year'}, inplace=True)

df.head(2).T
```

## Changes in Bechdel scores over time

Plot the change in the average Bechdel score over time. Scale the size of the line by the (logged) number of movies released that year.

```python
# Do some fancy re-indexing and groupby operations to get 
# average scores and number of movies between 1912 and 2013
avg_bechdel = df.set_index('Released_x').groupby(lambda x:x.year)['new_rating'].agg(np.mean).ix[1912:]
num_movies = df.set_index('Released_x').groupby(lambda x:x.year)['new_rating'].agg(lambda x:np.power(len(x)+1,1/3.0)).ix[1912:]

# Changing line widths from: http://stackoverflow.com/questions/19862011/matplotlib-change-linewidth-on-line-segments-using-list
coords = dict(avg_bechdel).items()
lines = [(start, end) for start, end in zip(coords[:-1], coords[1:])]
lines = LineCollection(lines, linewidths=dict(num_movies).values())

# Plot the figure
fig, ax = plt.subplots()
ax.add_collection(lines)
plt.autoscale()
plt.xlabel('Time',fontsize=18)
plt.ylabel('Avg. Bechdel score',fontsize=18)

```

### Simple linear extrapolation

We'll begin with a simple linear model of the change in the average Bechdel scores over time.

```python
avg_bechdel = df.set_index('Released_x').groupby(lambda x:x.year)['new_rating'].agg(np.mean).ix[1912:]
avg_bechdel.index.name = 'date'
avg_bechdel.name = 'bechdel'
avg_bechdel = avg_bechdel.reset_index()
```

```python
l_model = smf.ols(formula='bechdel ~ date', data=avg_bechdel).fit()
print l_model.summary()
```

```python
# Create a DataFrame "X" that we can extrapolate into
X = pd.DataFrame({"date": np.linspace(start=1912.,stop=2112.,num=201)})

# Plot the observed data
plt.plot(avg_bechdel['date'],avg_bechdel['bechdel'],c='g',label='Observed')

# Plot the predictions from the model
plt.plot(X,l_model.predict(X),c='r',label='Linear model',lw=2)
plt.legend(loc='lower right')
plt.ylim((0,3))
plt.xlim((1912,2112))
plt.xlabel('Year')
plt.ylabel('Avg. Bechdel Test')

plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

Re-arrange the `y=Bx+I` equation to solve for `y=30`: `x=(y-I)/B`

Then convert the float into a proper date and return the day of the week as well (Mondays are 0).

```python
year_est = (3 - l_model.params['Intercept'])/l_model.params['date']
year_int = int(year_est)
d = datetime.timedelta(days=(year_est - year_int)*365.25)
day_one = datetime.datetime(year_int,1,1)
date = d + day_one
print date, date.weekday()
```

Extrapolating this linear model forward in time, on **April 22, 2107**, the ***average*** movie will finally pass the Bechdel test. Just 93 years to go until that when magical Friday release when even the average summer blockbuster will have minimally-developed female characters! Hooray!

### Higher-order linear extrapolation

```python
q_model = smf.ols(formula='bechdel ~ date + I(date**2)', data=avg_bechdel).fit()

# Create a DataFrame "X" that we can extrapolate into
X = pd.DataFrame({"date": np.linspace(start=1912.,stop=2112.,num=201)})

# Plot the observed data
plt.plot(avg_bechdel['date'],avg_bechdel['bechdel'],c='g',label='Observed')

# Plot the predictions from the model
plt.plot(X,q_model.predict(X),c='r',label='Quadratic model',lw=2)
plt.legend(loc='lower right')
plt.ylim((0,3))
plt.xlim((1912,2112))
plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

This model says we're apparently at "Peak Bechdel" and a patriarchical cultural dystopia awaits us in the future.

### Time-series modeling

```python
# From: http://statsmodels.sourceforge.net/devel/examples/generated/ex_dates.html
avg_bechdel = df.set_index('Released_x').groupby(lambda x:x.year)['new_rating'].agg(np.mean).ix[1912:]
avg_bechdel.index = pd.to_datetime(avg_bechdel.index,format="%Y")
ar_model = sm.tsa.AR(avg_bechdel,freq='A').fit(maxlag=10, method='mle', disp=-1)
pred = ar_model.predict(start='1940-01-01', end='2030-01-01')
pred.plot('AR-model',lw=2,c='r'), avg_bechdel.plot(label='Observed',lw=1)
plt.legend(loc='lower right')

plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

```python
avg_bechdel.index = pd.to_datetime(avg_bechdel.index,format="%Y")
arma_model = sm.tsa.ARMA(avg_bechdel,(3,2),freq='A').fit()
pred = arma_model.predict(start='1940-01-01', end='2030-01-01')
pred.plot('ARMA-model',lw=2,c='r'), avg_bechdel.plot(label='Observed',lw=1)
plt.legend(loc='lower right')

plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

```python
avg_bechdel.index = pd.to_datetime(avg_bechdel.index,format="%Y")
arima_model = sm.tsa.ARIMA(avg_bechdel,(2,1,0),freq='A').fit()
pred = arima_model.predict(start='1940-01-01', end='2030-01-01')+avg_bechdel.mean()
pred.plot('ARIMA-model',lw=2,c='r'), avg_bechdel.plot(label='Observed',lw=1)
plt.legend(loc='lower right')

plt.xlabel('Year',fontsize=18)
plt.ylabel('Avg. Bechdel Test',fontsize=18)
```

## Relationship between Bechdel score and revenue

The first model we estimate is replicating the 538 analysis whether there's a relationship between revenue and Bechdel score, controlling for budget. We fit this model properly using `rating` as a categorical variable and logging the highly-skewed `Revenue` and `Budget` variables.

```python
m1 = smf.ols(formula='log(Adj_Revenue+1) ~ C(new_rating) + log(Adj_Budget+1)', data=df).fit()
print m1.summary()
```

```python
m1_b0 = np.exp(m1.params['Intercept'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b1 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.1.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b2 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.2.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))
m1_b3 = np.exp(m1.params['Intercept'] + m1.params['C(new_rating)[T.3.0]'] + m1.params['log(Adj_Budget + 1)']*np.log(np.mean(df['Adj_Budget'])))

print "Movies with an average budget and passing no element of the Bechdel test will make ${:,} on average".format(round(m1_b0,-3))
print "Movies with an average budget and passing the first element of the Bechdel test will make ${:,} on average".format(round(m1_b1,-3))
print "Movies with an average budget and passing the second element of the Bechdel test will make ${:,} on average".format(round(m1_b2,-3))
print "Movies with an average budget but passing all elements of the Bechdel test will make ${:,} on average".format(round(m1_b3,-3))
```

```python
# Add a jitter so the points aren't overlapping
x = df['new_rating'].apply(lambda x:float(x)+np.random.normal(0, 0.05))
y = df['Adj_Revenue']+1

#plt.plot(m1_bad_x['new_rating'],m1_bad.predict(m1_bad_x),c='r',lw=2,label='Model')

# Plot with an alpha so the overlaps reveal something about relative density
plt.scatter(x,y,alpha=.2,label='Data',c='g')
plt.xticks(arange(0,4))
plt.yscale('log')
plt.autoscale()
plt.xlabel('Bechdel score',fontsize=18)
plt.ylabel('Residual Adj. Revenue',fontsize=18)
#plt.ylim((-10,10))
plt.legend(loc='right',bbox_to_anchor=(1.275,.5))
```

Next we estimate a "bad" model using `rating` as a continuous variable so that we can visualize the model on top of the data itself.

```python
# Estimate model without the Bechdel test IV and containing only the IVs to get residulals
m1_bad_res = smf.ols(formula='log(Adj_Revenue+1) ~ log(Adj_Budget+1)', data=df).fit()

# Put these residual estimates in a temporary DataFrame
df_temp = df
df_temp['residuals'] = m1_bad_res.resid

# Estimate a bad model with a continuous Bechdel rating against the residuals
m1_bad = smf.ols(formula='residuals ~ new_rating', data=df).fit()
m1_bad_x = pd.DataFrame({'new_rating':np.linspace(min(df['new_rating']),max(df['new_rating']),100)})

# Add a jitter so the points aren't overlapping
x = df['new_rating'].apply(lambda x:float(x)+np.random.normal(0, 0.05))
y = df_temp['residuals']

plt.plot(m1_bad_x['new_rating'],m1_bad.predict(m1_bad_x),c='r',lw=2,label='Model')

# Plot with an alpha so the overlaps reveal something about relative density
plt.scatter(x,y,alpha=.2,label='Data',c='g')
plt.xticks(arange(0,4))
#plt.yscale('log')
plt.autoscale()
plt.xlabel('Bechdel score',fontsize=18)
plt.ylabel('Residual Adj. Revenue',fontsize=18)
plt.ylim((-10,10))
plt.legend(loc='right',bbox_to_anchor=(1.275,.5))
print "Every additional Bechdel 'point' adds {0} to the revenue and {1}".format(round(m1_bad.params['new_rating'],2),sig_tester(m1_bad.pvalues['new_rating']))
```
